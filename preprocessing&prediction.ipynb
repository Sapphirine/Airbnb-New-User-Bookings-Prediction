```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pandas import Series
from pandas import DataFrame

train_data = pd.read_csv('/Users/shuyi/2017Fall/BD/Project/proposal/airbnb/train_users_2.csv')
test_data = pd.read_csv('/Users/shuyi/2017Fall/BD/Project/proposal/airbnb/test_users.csv')
```

```python
train_data.head(10)
```

```python
train_data = train_data.drop(['date_account_created','timestamp_first_active'], axis=1)
test_data = test_data.drop(['date_account_created','timestamp_first_active'], axis=1)
```

```python
#print(train_data.at[3,'Year'])
def to_year(date):
    if date == date:
        return int(str(date)[:4])
    return date

def to_month(date):
    if date == date:
        return int(str(date)[5:7])
    return date

#print(train_data.at[3,'Year'])

train_data['Year'] = train_data['date_first_booking'].apply(to_year)
train_data['Month'] = train_data['date_first_booking'].apply(to_month)

#test_data['Year'] = test_data['date_first_booking'].apply(to_year)
#test_data['Month'] = test_data['date_first_booking'].apply(to_month)
```

```python
train_data['Year'].fillna(train_data['Year'].median (), inplace = True)
train_data['Month'].fillna(train_data['Month'].median(), inplace = True)
#test_data['Month'].fillna(test_data['Month'].median(), inplace = True)
#test_data['Year'].fillna(test_data['Year'].median(), inplace = True)
```

```python
train_data[['Year', 'Month']] = train_data[['Year', 'Month']].astype(int)
#test_data[['Year', 'Month']] = test_data[['Year', 'Month']].astype(int)

train_data.drop(['date_first_booking'], axis = 1, inplace = True)
```

```python
i = 0;
def get_gender(gender):
    global i
    i = i+1
    if gender != 'FEMALE' and gender != 'MALE':
        if (i%2):
            return 'FEMALE'
        else:
            return 'MALE'
    return gender

train_data['gender'] = train_data['gender'].apply(get_gender)
test_data['gender'] = test_data['gender'].apply(get_gender)
```

```python
train_data['gender'] = train_data['gender'].map({'FEMALE':1, 'MALE':0})
test_data['gender'] = test_data['gender'].map({'FEMALE':1, 'MALE':0})

```

```python
train_data['age'][train_data['age'] > 100] = np.NaN
test_data['age'][test_data['age'] > 100] = np.NaN

average_train = train_data['age'].mean()
std_train = train_data['age'].std()
count_train = train_data['age'].isnull().sum()

average_test = test_data['age'].mean()
std_test = test_data['age'].std()
count_test = test_data['age'].isnull().sum()

rand_train = np.random.randint(average_train - std_train, average_train + std_train, size = count_train)
rand_test = np.random.randint(average_test - std_test, average_test + std_test, size = count_test)

train_data['age'][np.isnan(train_data['age'])] = rand_train
test_data['age'][np.isnan(test_data['age'])] = rand_test

train_data['age'] = train_data['age'].astype(int)
test_data['age'] = test_data['age'].astype(int)
```

```python
affiliate_range = train_data['first_affiliate_tracked'].value_counts().index

count_aff_nan_train = train_data['first_affiliate_tracked'].isnull().sum()
count_aff_nan_test = test_data['first_affiliate_tracked'].isnull().sum()

rand_train1 = np.random.randint(0, len(affiliate_range), size = count_aff_nan_train)
rand_test1 = np.random.randint(0,len(affiliate_range), size = count_aff_nan_test)

train_data['first_affiliate_tracked'][train_data['first_affiliate_tracked'] != train_data['first_affiliate_tracked']] = affiliate_range[rand_train1]
test_data['first_affiliate_tracked'][test_data['first_affiliate_tracked'] != test_data['first_affiliate_tracked']] = affiliate_range[rand_test1]
```

```python
from sklearn import preprocessing

for attribute in train_data.columns:
    if attribute == 'country_destination' or attribute == 'id':
        continue
    if train_data[attribute].dtype == 'object':
        label = preprocessing.LabelEncoder()
        label.fit(np.unique(list(train_data[attribute].values) + list(test_data[attribute].values)))
        train_data[attribute] = label.transform(list(train_data[attribute].values))
        test_data[attribute] = label.transform(list(test_data[attribute].values))
        
```

```python
train = train_data.drop(['country_destination', 'id', 'Year', 'Month'], axis = 1)
train_label = train_data['country_destination']
test = test_data.drop(['id', 'date_first_booking'], axis = 1)
```

```python
train.head(10)
```

```python
country_dict = {'NDF':0, 'AU':1, 'CA':2, 'DE':3, 'ES':4, 'FR':5, 'GB':6, 'IT':7, 'NL':8, 'PT':9, 'US':10, 'other':11}
reverse_country_dict = {y:x for x,y in country_dict.items()}
train_label = train_label.map(country_dict)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import cross_validation
import xgboost as xgb
```

```python
#random forest
random_forest = RandomForestClassifier(n_estimators = 40)
random_forest.fit(train, train_label)
rf_prediction = random_forest.predict(test)
random_forest.score(train, train_label)
```

```python
rf_prediction = rf_prediction.astype(int)
rf_prediction = Series(rf_prediction).map(reverse_country_dict)
result_rf = DataFrame({'id':test_data['id'], 'country': rf_prediction})
result_rf = result_rf[(True-result_rf['country'].isin(['NDF']))]
result_rf.head(20)
```

```python
result_rf = DataFrame({'id': test_data['id'], 'country': rf_prediction})
result_rf.to_csv('prediction_rf.csv', index = False)
```

```python
#K-neighbors
K_neighbors = KNeighborsClassifier(n_neighbors = 4)
K_neighbors.fit(train, train_label)
k_prediction = K_neighbors.predict(test)
K_neighbors.score(train, train_label)
```

```python
k_prediction = k_prediction.astype(int)
k_prediction = Series(k_prediction).map(reverse_country_dict)
result_k = DataFrame({'id':test_data['id'], 'country': k_prediction})
result_k = result_k[(True-result_k['country'].isin(['NDF']))]
result_k.head(20)
```

```python
result_k = DataFrame({'id': test_data['id'], 'country': k_prediction})
result_k.to_csv('prediction_k.csv', index = False)
```

```python
#Gaussian Naive Bayes
clf = GaussianNB()
clf.fit(train, train_label)
gnb_prediction = clf.predict(test)
clf.score(train, train_label)
```

```python
gnb_prediction = gnb_prediction.astype(int)
gnb_prediction = Series(gnb_prediction).map(reverse_country_dict)
result_gnb = DataFrame({'id':test_data['id'], 'country': gnb_prediction})
result_gnb = result_gnb[(True-result_gnb['country'].isin(['NDF']))]
result_gnb.head(20)
```

```python
#Xgboost with default parameters
from xgboost.sklearn import XGBClassifier

xgb1 = XGBClassifier()
xgb1.fit(train, train_label)
xgb1.score(train, train_label)
```

```python
#Xgboost with adjusted parameters
xgb2 = XGBClassifier(max_depth=6, learning_rate = 0.3, n_estimators=25, objective='multi:softprob',subsample=0.5,colsample_bytree=0.5,seed = 0)
xgb2.fit(train, train_label)
xgb2_prediction = xgb2.predict(test)
xgb2.score(train, train_label)
```

```python
#Xgboost with adjusted parameters and cross validation
from sklearn.cross_validation import train_test_split
t, v = train_test_split(train_data, test_size = 0.3, random_state = 1)
xgb_train = t.drop(['country_destination', 'id', 'Year', 'Month'], axis = 1)
xgb_train_label = t['country_destination']
val = v.drop(['country_destination', 'id', 'Year', 'Month'], axis = 1)
val_label = v['country_destination']
xgb_train_label = xgb_train_label.map(country_dict)
val_label = val_label.map(country_dict)

train_xgb = xgb.DMatrix(xgb_train, xgb_train_label)
val_xgb = xgb.DMatrix(val, val_label)
test_xgb = xgb.DMatrix(test)

train_all = xgb.DMatrix(train)
train_label_all = xgb.DMatrix(train_label)
```

```python
params={
'booster':'gbtree',
'objective': 'multi:softmax', 
'num_class':12, 
'gamma':0.1,  
'max_depth':12, 
'lambda':2,  
'subsample':0.7, 
'colsample_bytree':0.7, 
'min_child_weight':3, 
'silent':0 ,
'eta': 0.007, 
'seed':1000,
'nthread':7,
}
plst = list(params.items())
num_rounds = 5000 
watchlist = [(train_xgb, 'train'),(val_xgb, 'val')]
model = xgb.train(plst, train_xgb, num_rounds, watchlist, early_stopping_rounds=400)
```

```python
from sklearn.metrics import accuracy_score
prediction_train = model.predict(train_all)
xgb_score = accuracy_score(prediction_train, train_label)
print(xgb_score)
```
